```python
def neg_sum(array: list):
    
    min_val = min(array) # Find min and max values positions first
    max_val = max(array)
    min_val_pos = array.index(min_val)
    max_val_pos = array.index(max_val)
    
    if min_val_pos > max_val_pos: # Swap min and max if min bigger than max
        min_val_pos, max_val_pos = max_val_pos, min_val_pos
    
    min_max_array = array[min_val_pos + 1 : max_val_pos + 1]
    sum = 0
    
    if min_max_array: # Check if the slice is not empty
        for elem in min_max_array: # Iterate through the slice getting the sum
            if elem < 0:
                sum += elem
    return sum


test_array = [1, 3, -10, 8, 0, 10, -2, 5, -4, 10, -3, 15]
print(neg_sum(test_array))
```
